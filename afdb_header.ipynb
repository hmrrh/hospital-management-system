```python
# Loop through each record name
for record_name in record_names:
    # Read the header file
    header = wf.rdheader(os.path.join(path, record_name))

    # Print all attributes in the header object
    print(f"Record: {record_name}")
    print("Header attributes:")
    for key, value in header.__dict__.items():
        print(f"{key}: {value}")
    print()
```
